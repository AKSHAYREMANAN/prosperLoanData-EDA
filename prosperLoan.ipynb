15. REMOVE OUTLIEARS

In Exploratory Data Analysis (EDA), removing outliers is an important part of the process. Outliers are data points that are significantly different from the majority of the data points in a dataset. They can affect the calculated parameters and skew the results of statistical . There are several methods to detect and remove outliers

```python

Q1 = df['Investors'].quantile(0.25)
Q3 = df['Investors'].quantile(0.99)
IQR = Q3 - Q1
print(IQR)
((df['Investors'] < (Q1 - 1.5 * IQR)) | (df['Investors'] > (Q3 + 1.5 * IQR))).sum()
```

```python
sns.boxplot(df['Investors'])
```

```python
Q1 = df['Recommendations'].quantile(0.25)
Q3 = df['Recommendations'].quantile(0.95)
IQR = Q3 - Q1
print(IQR)
((df['Recommendations'] < (Q1 - 1.5 * IQR)) | (df['Recommendations'] > (Q3 + 1.5 * IQR))).sum()
```

20. PIE CHART

Pie charts are commonly used in EDA to visualize the proportion or percentage of different categories that make up a whole. They are particularly useful when there are only a few categories, and the goal is to show the relative size of each category.

```python
loan_data = pd.DataFrame({
    'Lp_servicefees': [100, 150, 120, 170],
    'Lp_collectionfees': [50, 75, 60, 80],
    'Lp_interestandfees': [75, 110, 90, 130],
    'Lp_grossprincipalloss': [125, 175, 140, 200]
})

# Calculate the percentage distribution
percentage_distribution = loan_data[['Lp_servicefees', 'Lp_collectionfees', 'Lp_interestandfees', 'Lp_grossprincipalloss']].sum() / loan_data[['Lp_servicefees', 'Lp_collectionfees', 'Lp_interestandfees', 'Lp_grossprincipalloss']].sum().sum() * 100

# Create a pie chart
fig, ax = plt.subplots()
ax.pie(percentage_distribution, labels=['Lp_servicefees', 'Lp_collectionfees', 'Lp_interestandfees', 'Lp_grossprincipalloss'], autopct='%1.1f%%')
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Fee and Loss Distribution in Loan Data')
plt.show()
```
